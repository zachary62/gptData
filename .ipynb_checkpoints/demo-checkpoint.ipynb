## Pivot/Unpivot

````python
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage, AIMessage
openai_api_key='...'
chat = ChatOpenAI(temperature=0, openai_api_key=openai_api_key)
````

````python
from langchain import PromptTemplate

template = """
Answer the following questions as best you can. 

Are there any columns to unpivot? After unpivot, each numerical row should have the same granularity, and suming the column should provide meaningful value.

Use the following format:

Dataframe file name:
Dataframe descroption:
Dataframe sample:
Need unpivot: Yes/No
Unpivot: ```df = pd.melt(df,...)``` (must give the code if need unpivot)

Begin!

Dataframe file name: {name}
Dataframe descroption: {description}
Dataframe sample: {sample}
Need unpivot:"""

prompt = PromptTemplate(
    input_variables=["name", "description", "sample"],
    template=template,
)
````

````python
import pandas as pd
df = pd.read_csv("./data/Metro_mlp_uc_sfrcondo_sm_month.csv")
df
````

````python
final_prompt = prompt.format(name='Metro_mlp_uc_sfrcondo_sm_month', description='Median List Price: The median price at which homes across various geographies were listed.', sample=str(df.sample(5).to_string()))
print(final_prompt)
````

````python
result = chat(
    [
        SystemMessage(content="You are a nice AI bot that follows the instructions to clean datasets."),
        HumanMessage(content=final_prompt)
    ]
)
````

````python
def find_first_yes_or_no(my_string):
    # find the first occurrence of "yes" and "no" in the string
    index_yes = my_string.find("Yes")
    index_no = my_string.find("No")

    # determine which one comes first in the string
    if index_yes == -1 and index_no == -1:
        return None
    elif index_yes == -1:
        return False
    elif index_no == -1:
        return True
    elif index_yes < index_no:
        return True
    else:
        return False
    
import re

def extract_python_code(input_string):
    pattern = r"Unpivot: ```(.*?)```"
    matches = re.findall(pattern, input_string, re.DOTALL)
    if len(matches) > 0:
        return matches[0].strip()
    else:
        return None
````

````python
find_first_yes_or_no(result.content)
````

````python
codes = extract_python_code(result.content)
codes
````

````python
exec(codes)
df
````

# Data discovery

Compared to embedding:
- Embedding need a (hard-to-specify and context-specific) threshold to tell if matched or not. We can decide such threshold by gpt (e.g., binary search)
- We need to decide how to cluster attribuets (e.g., month, year, day as date for join key). It is still hard to specify threshold to clustering.
- ... There must be other limitations
- We call write python codes for efficiency. Call external services.
- Make GPT agent. It can further asks for eda to increase estimation. E.g., two columns are likely to match. Ask to check their statistics.

````python
template = """
Answer the following questions as best you can. 

Given two datsets and column description, are there columns that reference the same entity and can be the matched (with appropriate transformation)?

Follow strictly the following format:

Database description:
Table1 description:
Table1 sample:
Table2 description:
Table2 sample:
Matching Attributes: ```[("table1 attribute", "table2 attribute"),...]```

Begin!

Database description: {description0}
Table1 description: {description1}
Table1 sample: {sample1}
Table2 description: {description2}
Table2 sample: {sample2}
Matching Attributes:"""

prompt = PromptTemplate(
    input_variables=["description0", "description1", "description2", "sample1", "sample2"],
    template=template,
)
````

````python
import pandas as pd
df1 = pd.read_csv("./data/baseball/fielding.csv")
df2 = pd.read_csv("./data/baseball/manager_half.csv")
````

````python
final_prompt = prompt.format(
    description0="""
    """,
    description1="""""", 
    description2="""""", 
    sample1=str(df1.sample(5).to_string()),
    sample2=str(df2.sample(5).to_string())
                            )
print(final_prompt)
````

````python
chat = ChatOpenAI(temperature=0, openai_api_key=openai_api_key)
result = chat(
    [
        SystemMessage(content="You are a nice AI bot that follows trictly the format from the prompt."),
        HumanMessage(content=final_prompt)
    ]
)
````
